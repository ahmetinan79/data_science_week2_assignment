```python
import numpy as np
import pandas as pd
```

## 1. Tum ogrencilerin bilgilerini tek bir dataframe icinde su sutunlardan olusacak sekilde birlestiriniz; " Sinif ismi, Ogrenci Adi, Dogru Sayisi, Yanlis Sayisi, Bos Sayisi".

```python
#Py_mind
pymind = pd.read_excel('py_mind.xlsx', sheet_name=None)  # We take all the sheets from the file.

pymind1 = pymind.copy()  # We create a copy of the file to keep the original file intact.

names1=[]
for i in pymind1.keys():  # We get the list of the students from each sheet in the file. 
    names1.append(pymind1[i].columns[0])

df1 = pd.DataFrame({'deneme':[0]})  # We create a dataframe to add our info from our dataframes. 

for i in pymind1.keys():  # We merge the info in each dataframe as columns. 
    df1 = pd.concat([df1,pymind1[i].T],ignore_index=True)  # We transpose the columns to get the desired format of data.

df1.drop(columns=[i for i in range(20)]+['deneme'],inplace=True)  # We drop the unnecessary columns. 

list1=[i for i in range(4,21,2)]+[0,1,3]  
df1.drop(list1,axis=0,inplace=True)  # We drop the unnecessary rows.

df1['Ogrenci_Adi'] = names1   # We add the student names as a new column.
df1['Sinif_Ismi'] = "Py_mind"  # We add the class name as a new column. 
df1.columns = ['Dogru_Sayisi','Yanlis_Sayisi', 'Bos_Sayisi', 'Ogrenci_Adi', 'Sinif_Ismi']  # We rename the columns.
df1 = df1[['Sinif_Ismi','Ogrenci_Adi','Dogru_Sayisi','Yanlis_Sayisi','Bos_Sayisi']]  # We rearrange the columns.
new_index1 = np.arange(1,(len(df1.Ogrenci_Adi))+1)  # We reset the index numbers.
df1.set_index(new_index1, inplace=True)
df1
```

```python
#Py_opinion
pyopinion = pd.read_excel('py_opinion.xlsx', sheet_name=None)

pyopinion1 = pyopinion.copy()

names2=[]
for i in pyopinion1.keys():
    names2.append(pyopinion1[i].columns[0])

df2 = pd.DataFrame({'deneme':[0]})

for i in pyopinion1.keys():
    df2 = pd.concat([df2,pyopinion1[i].T],ignore_index=True)

df2.drop(columns=[i for i in range(20)]+['deneme'],inplace=True)

list2=[i for i in range(1,20,2)]+[0]  
df2.drop(list2,axis=0,inplace=True)
df2['Ogrenci_Adi'] = names2
df2['Sinif_Ismi'] = "Py_opinion"
df2.columns = ['Dogru_Sayisi','Yanlis_Sayisi', 'Bos_Sayisi', 'Ogrenci_Adi', 'Sinif_Ismi']
df2 = df2[['Sinif_Ismi','Ogrenci_Adi','Dogru_Sayisi','Yanlis_Sayisi','Bos_Sayisi']]
new_index2 = np.arange(1,(len(df1.Ogrenci_Adi))+1)
df2.set_index(new_index1, inplace=True)
df2
```

```python
#Py_science
pyscience = pd.read_excel('py_science.xlsx', sheet_name=None)

pyscience1 = pyscience.copy()

names3=[]
for i in pyscience1.keys():
    names3.append(pyscience1[i].columns[0])

df3 = pd.DataFrame({'deneme':[0]}) 

for i in pyscience1.keys():
    df3 = pd.concat([df3,pyscience1[i].T],ignore_index=True)

df3.drop(columns=[i for i in range(20)]+['deneme'],inplace=True)

list3=[i for i in range(1,16,2)]+[0]
df3.drop(list3,axis=0,inplace=True)

df3['Ogrenci_Adi'] = names3
df3['Sinif_Ismi'] = "Py_science"
df3.columns = ['Dogru_Sayisi','Yanlis_Sayisi', 'Bos_Sayisi', 'Ogrenci_Adi', 'Sinif_Ismi']
df3 = df3[['Sinif_Ismi','Ogrenci_Adi','Dogru_Sayisi','Yanlis_Sayisi','Bos_Sayisi']]
df3.fillna('x',inplace=True)  # We replace NaN values with x to locate and correct them. 
df3.replace({'Dogru_Sayisi':{4:15},'Yanlis_Sayisi':{'x':1}, 'Bos_Sayisi':{'x':4}},inplace=True) # We correct wrong values.
new_index3 = np.arange(1,(len(df3.Ogrenci_Adi))+1)
df3.set_index(new_index3, inplace=True)
df3
```

```python
#Py_sense
pysense = pd.read_excel('py_sense.xlsx', sheet_name=None)

pysense1 = pysense.copy()

names4=[]
for i in pysense1.keys():
    names4.append(pysense1[i].columns[0])

df4 = pd.DataFrame({'deneme':[0]})

for i in pysense1.keys():
    df4 = pd.concat([df4,pysense1[i].T],ignore_index=True)

df4.drop(columns=[i for i in range(20)]+['deneme'],inplace=True)

list4=[i for i in range(1,20,2)]+[0]  
df4.drop(list4,axis=0,inplace=True)

df4['Ogrenci_Adi'] = names4
df4['Sinif_Ismi'] = "Py_sense"
df4.columns = ['Dogru_Sayisi','Yanlis_Sayisi', 'Bos_Sayisi', 'Ogrenci_Adi', 'Sinif_Ismi']
df4 = df4[['Sinif_Ismi','Ogrenci_Adi','Dogru_Sayisi','Yanlis_Sayisi','Bos_Sayisi']]
new_index4 = np.arange(1,(len(df4.Ogrenci_Adi))+1)
df4.set_index(new_index1, inplace=True)
df4
```

```python
#Compiled data of classes
df_compiled = pd.concat([df1,df2,df3,df4],ignore_index=True)  # We compile 4 dataframes.
# We turn the values in 3 columns from object to integer. 
df_compiled[['Dogru_Sayisi','Yanlis_Sayisi','Bos_Sayisi']]=df_compiled[['Dogru_Sayisi','Yanlis_Sayisi','Bos_Sayisi']].astype(int)
new_index = np.arange(1,(len(df_compiled.Ogrenci_Adi))+1)  # We reset the index values. 
df_compiled.set_index(new_index, inplace=True)
df_compiled
```
